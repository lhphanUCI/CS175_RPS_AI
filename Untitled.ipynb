```python
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt

sess = tf.InteractiveSession()


def crop_image(img, crop):
    return tf.image.crop_to_bounding_box(img,
                                        crop[0],
                                        crop[1],
                                        crop[2],
                                        crop[3])

# img
img1 = np.random.random([128, 128, 3])
img2 = np.random.random([128, 128, 3])
img3 = np.random.random([128, 128, 3])
images = np.array([img1, img2, img3])
images = tf.convert_to_tensor(images)


print (images.shape)

# crop value
offsetHeight = np.random.randint(0, 64)
offsetWidth = np.random.randint(0, 64)
crop_value = [offsetHeight, offsetWidth, 64, 64]
crop_values = []
for i in range(images.shape[0]):
    crop_values.append(crop_value)
crop_values = tf.convert_to_tensor(crop_values)




fn = lambda x: crop_image(x[0], x[1])
elems = (images, crop_values)
cropped_images = tf.map_fn(fn, elems=elems, dtype=tf.float64)
result = sess.run(cropped_images)

print (result.shape)
plt.imshow(result[0])
plt.show()
```
